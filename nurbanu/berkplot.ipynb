```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from itertools import cycle

mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'

mpl.rcParams['mathtext.fontset'] = 'custom'
mpl.rcParams['mathtext.it'] = 'Arial:italic'
mpl.rcParams['mathtext.rm'] = 'Arial'

%matplotlib inline
```

```python
df = pd.read_csv('berk_data_en.csv')
```

```python
countries = df.country.unique()
countries
```

```python
df[df['country'] == countries[1]]
```

```python
"./figures/" + countries[0].lower() + ".svg"
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(df[df['country'] == countries[0]].Year, np.log(df[df['country'] == countries[0]]['Health Expenditure']), label=f"Saglik Harcamalari", linewidth=3)
ax.plot(df[df['country'] == countries[0]].Year, np.log(df[df['country'] == countries[0]]['GDP']), label=f"Gayrisafi Yurtiçi Hasıla", linewidth=3)
ax.plot(df[df['country'] == countries[0]].Year, np.log(df[df['country'] == countries[0]]['Suicide Rate']), label=f"İntihar Oranı", linewidth=3)
ax.set_title(countries[0], fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='both', which='minor', labelsize=18)
plt.legend(loc="best", fontsize=20)
plt.tight_layout()
plt.savefig("./figures/" + countries[0].lower() + ".svg", bbox_inches="tight", format="svg")
```

```python
for i in range(len(countries)):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.plot(df[df['country'] == countries[i]].Year, np.log(df[df['country'] == countries[i]]['Health Expenditure']), label=f"Saglik Harcamalari", linewidth=3)
    ax.plot(df[df['country'] == countries[i]].Year, np.log(df[df['country'] == countries[i]]['GDP']), label=f"Gayrisafi Yurtiçi Hasıla", linewidth=3)
    ax.plot(df[df['country'] == countries[i]].Year, np.log(df[df['country'] == countries[i]]['Suicide Rate']), label=f"İntihar Oranı", linewidth=3)
    ax.set_title(countries[i], fontsize=20)
    ax.tick_params(axis='both', which='major', labelsize=20)
    ax.tick_params(axis='both', which='minor', labelsize=18)
    plt.legend(loc="best", fontsize=20)
    plt.tight_layout()
    plt.savefig("./figures/" + countries[i].lower() + ".svg", bbox_inches="tight", format="svg")
    plt.close()
```
